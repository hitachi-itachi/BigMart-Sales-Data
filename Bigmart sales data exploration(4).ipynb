```python
stats_alive = bigmart_sales.groupby('Item_Type')['Item_Outlet_Sales'].describe().reset_index()[['Item_Type','25%','75%']] #shrink em to 25 percentile and 75 percentile
stats_alive
```

```python
stats_Visibility = bigmart_sales.groupby('Item_Type')['Item_Visibility'].describe().reset_index()[['Item_Type','25%','75%']] #shrink em to 25 percentile and 75 percentile
stats_Visibility
```

```python
stats_alive.columns = stats_alive.columns.str.replace('25%','Item_Outlet_Sales_25%')
stats_alive.columns = stats_alive.columns.str.replace('75%','Item_Outlet_Sales_75%')
stats_alive
```

```python
stats_Visibility.columns = stats_Visibility.columns.str.replace('25%','Item_Visiblity_25%')
stats_Visibility.columns = stats_Visibility.columns.str.replace('75%','Item_Visbility_75%')
stats_Visibility
```

```python
#created bigmart subset 

bigmart_sales_subset = bigmart_sales[["Item_Visibility","Item_Type","Item_MRP","Outlet_Size","Outlet_Location_Type","Outlet_Type", "Item_Outlet_Sales"]]
```

```python
#stats_alive = bigmart_sales_subset.groupby('Item_Type')['Item_Outlet_Sales'].describe().reset_index()[['Item_Type','Item_Outlet_Sales_25%','Item_Outlet_Sales_75%']]
#stats_alive
```

```python
bigmart_sales_subset = bigmart_sales_subset.merge(stats_alive, on ='Item_Type', how = 'left')
bigmart_sales_subset
```

```python
bigmart_sales_subset = bigmart_sales_subset.merge(stats_Visibility, on ='Item_Type', how = 'left')
bigmart_sales_subset
```

```python
mask = bigmart_sales_subset['Item_Outlet_Sales']<bigmart_sales_subset['Item_Outlet_Sales_25%']
bigmart_sales_subset['Item_Outlet_Sales'] = bigmart_sales_subset['Item_Outlet_Sales_25%']

mask = bigmart_sales_subset['Item_Outlet_Sales']>bigmart_sales_subset['Item_Outlet_Sales_75%']
bigmart_sales_subset['Item_Outlet_Sales'] = bigmart_sales_subset['Item_Outlet_Sales_75%']
```

```python
mask = bigmart_sales_subset['Item_Visibility']<bigmart_sales_subset['Item_Visiblity_25%']
bigmart_sales_subset['Item_Visibility'] = bigmart_sales_subset['Item_Visiblity_25%']

mask = bigmart_sales_subset['Item_Visibility']>bigmart_sales_subset['Item_Visbility_75%']
bigmart_sales_subset['Item_Visibility'] = bigmart_sales_subset['Item_Visbility_75%']
```
